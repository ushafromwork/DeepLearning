```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
#loading the data frame from sklearn datsets
breast_cancer_dataset=sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
data_frame=pd.DataFrame(breast_cancer_dataset.data,columns=breast_cancer_dataset.feature_names)
data_frame.head()
```

```python
data_frame['label']=breast_cancer_dataset.target
```

```python
x=data_frame.drop(columns='label',axis=1)
y=data_frame['label']
```

```python
print(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
print(x.shape,x_train.shape,x_test.shape)
```

STanrdize the data to get better accuracy

```python
from sklearn.preprocessing import StandardScaler
```

```python
scalar=StandardScaler()
```

```python
x_train_std=scalar.fit_transform(x_train)
x_test_std=scalar.transform(x_test) #we shouldn't fit test data
```

**Building the Neural Network**

image.png


```python
#importing tensorflow and keras
import tensorflow as tf
tf.random.set_seed(3)#While training NN we need several parameters initialized, each time we train NN,the accuracy value may change every time we run,
#to avoid that we set set_seed
from tensorflow import keras #tensorflow is a deeplearning library by google
```

```python
#Setting up the layers of Neural Network---input--hidden--output
model=keras.Sequential([
                         keras.layers.Flatten(input_shape=(30,)),    # to conver data into single dimensional array,,input layer
                         keras.layers.Dense(20,activation='relu'),   # hidden layer with 20 neurons
                         keras.layers.Dense(2,activation='sigmoid')  #2=number of outputs(classification) either 0 or 1 Output layer
])
```

```python
#compiling the neural network
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

**Training the Neural Network**

```python
history = model.fit(x_train_std,y_train,validation_split=0.1,epochs=10)#epoch is no.of times it executes
```

Visualizing accuracy and loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.legend(['training data','validation data'], loc ='lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend(['training data','validation data'], loc ='lower right')
```

**Accuracy of the model on test data**

```python
loss,accuracy=model.evaluate(x_test_std,y_test,)
print(accuracy)
```

```python
print(x_test_std.shape)
print(x_test_std[0])
```

Build predictive system based on trained model

```python
y_pred=model.predict(x_test_std)
print(y_pred.shape)
print(y_pred[0])
```

model.predict() gives the prediction probability of each class for that data point

```python
#argmax function

my_list = [0.6,0.76]

index_of_max_value=np.argmax(my_list)
print(my_list)
print(index_of_max_value)
```

```python
y_pred_labels=[np.argmax(i) for i in y_pred] #1--2nd value maximum,0--1st value maximum
print(y_pred_labels)
```

**Building the predictive system**

```python
input_data=(13.54,14.36,87.46,566.3,0.09779,0.08129,0.06664,0.04781,0.1885,0.05766,0.2699,0.7886,2.058,23.56,0.008462,0.0146,0.02387,0.01315,0.0198,0.0023,15.11,19.26,99.7,711.2,0.144,0.1773,0.239,0.1288,0.2977,0.07259)

#Change the input_data to numpy array
input_as_array=np.asarray(input_data)

input_reshaped=input_as_array.reshape(1,-1)# As we are predicting for one data point

input_std=scalar.transform(input_reshaped)

prediction=model.predict(input_std)
print(prediction)

prediction_label=[np.argmax(prediction)]
print(prediction_label)

if(prediction_label[0] == 0):
  print('Tumor is Malignant')

else:
  print('Tumor is Benign')
```
