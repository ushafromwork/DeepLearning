Resizing image using pillow library

all images should be in same size to understand by neural networks

```python

from PIL import Image
```

```python
img_resized.save('sunflower_image_resized.jpg')
```

**Convert RGB images into Grayscale images using openCV**

```python
import cv2
```

```python
img=cv2.imread('/content/Sunflower-field-Fargo-North-Dakota.jpg')
```

```python
type(img)
```

```python
print(img.shape)
```

```python
grayscale_image=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
```

```python
type(grayscale_image)
```

```python
print(grayscale_image.shape)
```

```python
#cv2.imshow() will display the image but it is not allowed in colab
```

```python
from google.colab.patches import cv2_imshow
```

```python
#displaying the image
cv2_imshow(grayscale_image)
```

```python
#saving the grayscale image
cv2.imwrite('sunflower_grayscale_image.jpg',grayscale_image)
```
